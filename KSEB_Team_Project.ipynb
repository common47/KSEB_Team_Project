```python
from ultralytics import YOLO

# Load a pretrained YOLOv10n model
# model = YOLO("yolov12n.pt")
model = YOLO("yolo11n.pt")
imgPath = "/content/Small-Fire-Smoke-OD-Dataset-6/data.yaml"
model.train(data=imgPath,
             epochs=50,
             patience=10,
             batch=32)
```

```python
import cv2

# ì˜ìƒ íŒŒì¼ ê²½ë¡œ
video_path = '/content/í™”ìž¬.mp4'
# ìµœì¢…ì ìœ¼ë¡œ ì €ìž¥í•  ë™ì˜ìƒ íŒŒì¼ ê²½ë¡œ
output_video_path = '/content/yolo_test_video_reconstructed2.mp4'
fps = 30  # ì €ìž¥í•  ë™ì˜ìƒì˜ í”„ë ˆìž„ ì†ë„ (FPS)

# ë¹„ë””ì˜¤ ìº¡ì²˜ ê°ì²´ ìƒì„±
capture = cv2.VideoCapture(video_path)

# í”„ë ˆìž„ í¬ê¸° ê°€ì ¸ì˜¤ê¸°
frame_width = int(capture.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(capture.get(cv2.CAP_PROP_FRAME_HEIGHT))

# ë¹„ë””ì˜¤ ë¼ì´í„° ê°ì²´ ìƒì„±
fcc = cv2.VideoWriter_fourcc(*'mp4v')
video_writer = cv2.VideoWriter(output_video_path, fcc, fps, (frame_width, frame_height))

# í”„ë ˆìž„ì„ í•˜ë‚˜ì”© ì½ì–´ì™€ì„œ ë°”ë¡œ ìƒˆë¡œìš´ MP4 íŒŒì¼ì— ì €ìž¥
while capture.isOpened():
    ret, frame = capture.read()
    if ret:
        detection = model(frame)[0]
        boxInfos = detection.boxes.data.tolist()
        for i in range(0, len(boxInfos)):
            data = boxInfos[i]
            # data : [xmin, ymin, xmax, ymax, confidence_score, class_id]
            # ë°”ìš´ë”©ë°•ìŠ¤ ì‚¬ê°í˜•ì„ ê·¸ë¦¬ê¸° ìœ„í•´ì„œ ì¢Œí‘œ ê°’ ë¶ˆëŸ¬ì˜¤ê¸°
            xmin, ymin, xmax, ymax = int(data[0]), int(data[1]), int(data[2]), int(data[3])
            # ë°”ìš´ë”©ë°•ìŠ¤ ì‚¬ê°í˜• ê·¸ë¦¬ê¸°
            cv2.rectangle(frame, (xmin, ymin), (xmax, ymax), (0,255,0), 2)

            #í´ëž˜ìŠ¤ëª… ë¶ˆëŸ¬ì˜¤ê¸°
            label = int(data[5])
            classNames = model.names
            classLabel = classNames[label]
            confidence = str(round(data[4],2))
            textValue = classLabel+'_' + confidence + '%'
            #í´ëž˜ìŠ¤ëª… ë° confidence ë¶™ì´ê¸°
            cv2.putText(frame, textValue, (xmin, ymin), cv2.FONT_ITALIC, 1, (255,255,255), 2)
            # í”„ë ˆìž„ ì €ìž¥
        # í”„ë ˆìž„ì„ ë¹„ë””ì˜¤ì— ì €ìž¥
        video_writer.write(frame)
    else:
        break

# ë¹„ë””ì˜¤ ìº¡ì²˜ ë° ë¼ì´í„° ê°ì²´ í•´ì œ
capture.release()
video_writer.release()

print(f"Video reconstructed and saved as {output_video_path}")
```

```python
import cv2
import pyttsx3

# TTS ì—”ì§„ ì´ˆê¸°í™”
engine = pyttsx3.init()
is_fire_announced = False  # TTS ì¤‘ë³µ ë°©ì§€ìš© í”Œëž˜ê·¸

# ì˜ìƒ íŒŒì¼ ê²½ë¡œ
video_path = '/content/í™”ìž¬.mp4'
output_video_path = '/content/yolo_test_video.mp4'
fps = 30

# ë¹„ë””ì˜¤ ìº¡ì²˜ ê°ì²´ ìƒì„±
capture = cv2.VideoCapture(video_path)
frame_width = int(capture.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(capture.get(cv2.CAP_PROP_FRAME_HEIGHT))

# ë¹„ë””ì˜¤ ë¼ì´í„° ê°ì²´ ìƒì„±
fcc = cv2.VideoWriter_fourcc(*'mp4v')
video_writer = cv2.VideoWriter(output_video_path, fcc, fps, (frame_width, frame_height))

# í”„ë ˆìž„ ì²˜ë¦¬ ë£¨í”„
while capture.isOpened():
    ret, frame = capture.read()
    if not ret:
        break

    detection = model(frame)[0]
    boxInfos = detection.boxes.data.tolist()
    for data in boxInfos:
        xmin, ymin, xmax, ymax = int(data[0]), int(data[1]), int(data[2]), int(data[3])
        confidence = data[4]
        label = int(data[5])

        # confidenceê°€ 0.5 ì´ìƒì¼ ë•Œë§Œ ì²˜ë¦¬
        if confidence >= 0.5:
            classNames = model.names
            classLabel = classNames[label]
            textValue = f"{classLabel}_{round(confidence * 100, 2)}%"

            # ë°”ìš´ë”©ë°•ìŠ¤ ê·¸ë¦¬ê¸°
            cv2.rectangle(frame, (xmin, ymin), (xmax, ymax), (0, 255, 0), 2)
            cv2.putText(frame, textValue, (xmin, ymin), cv2.FONT_ITALIC, 1, (255, 255, 255), 2)

    # í”„ë ˆìž„ì„ ë¹„ë””ì˜¤ì— ì €ìž¥
    video_writer.write(frame)

# ìžì› í•´ì œ
capture.release()
video_writer.release()

print(f"Video reconstructed and saved as {output_video_path}")
```
